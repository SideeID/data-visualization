```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')

# Define data
MataKuliah = ['Matematika', 'Fisika', 'Kimia', 'Komputer', 'Bahasa']
Nilai = [80, 96, 65, 79, 82]

# Create the pie chart
ax.pie(Nilai, labels=MataKuliah, autopct="%1.2f%%")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')

x = np.linspace(0, 8, 100)
y = 4 + 2 * np.sin(2 * x)

fig, ax = plt.subplots()
ax.plot(x, y, linewidth=2.8)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(-2, 8), yticks=np.arange(-2, 8))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')

np.random.seed(3)
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

fig, ax = plt.subplots()
ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')

np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

fig, ax = plt.subplots()
scatter = ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

cbar = plt.colorbar(scatter)
cbar.set_label('Color Scale')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4 * x / 8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2 * x / 8 + np.random.uniform(0.0, 0.5, len(x))

fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=0.5, linewidth=0)

ax.plot(x, (y1 + y2) / 2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
